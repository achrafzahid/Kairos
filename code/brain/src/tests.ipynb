```python
import os
from alpaca.data.historical import CryptoHistoricalDataClient
from alpaca.data.requests import CryptoBarsRequest
from alpaca.data.timeframe import TimeFrame
from datetime import datetime

# Load the .env file into the system environment

# Use os.getenv() to retrieve them
api_key = os.getenv("APCA_API_KEY_ID")
secret_key = os.getenv("APCA_API_SECRET_KEY")

# 1. Initialize the Client with your credentials
client = CryptoHistoricalDataClient(api_key, secret_key)

# 2. Creating request object
request_params = CryptoBarsRequest(
    symbol_or_symbols=["BTC/USD"],
    timeframe=TimeFrame.Day,
    start=datetime(2022, 9, 1),
    end=datetime(2022, 9, 7)
)

# 3. Retrieve data using the client
btc_bars = client.get_crypto_bars(request_params)

# 4. Convert to dataframe
print(btc_bars.df.reset_index().head())
```
